## idependent dataset處理

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("/content/drive/MyDrive/Colab_Notebooks/test.csv")
# 之後要寫的fasta檔案
independent_predata = open("/content/drive/MyDrive/Colab_Notebooks/independent_predata.txt", "w")
data
```

```python
each_id = data["sample id"]
whole_seq = data["sequence"]
```

```python
n = int(15)
length = int(n*2+1)
```

```python
# 抓K的位子，且長度為31
k_id = []
k_index = []
k_seq = []
for i in range(0,len(whole_seq)):
    each_seq = whole_seq[i]
    for pos in range(0,len(each_seq)):
        if each_seq[pos] == "K" and len(each_seq[pos-n:pos+n+1])== length :
            independent_predata.write('>' + str(each_id[i]) + '_' + str(pos+1) + "\n" + each_seq[pos-n:pos+n+1]+ "\n")
            k_id.append(each_id[i])
            k_index.append(pos+1)
            k_seq.append(each_seq[pos-n:pos+n+1])
```

```python
# 偷看抓K的結果
new_data = pd.DataFrame({
    "k_id": k_id,
    "k_index": k_index,
    "k_seq": k_seq
})
new_data
```

```python
independent_predata.close()
```
