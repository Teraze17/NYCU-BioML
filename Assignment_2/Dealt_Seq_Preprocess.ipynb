# 抓 positive 位置 Ｋ

```python
# ppp=positive position p  因為數據中是以1開始,故在這邊運行時需要將位置-1才會是真正的positive K(從0開始)
ppp = Ub[1]
pp=[]

for i in range(len(ppp)):
    pp.append(ppp[i] - 1)

print(pp)
```

```python
sequence = Ub[3]
segment = []

for i in range(15000):
  selected_sequence = sequence[i]
  selected_pp = pp[i]

  if selected_pp < 16:
    start_index = 0
  else:
    start_index = selected_pp - 15

  end_index = min(selected_pp + 16, len(selected_sequence))

  segments = selected_sequence[start_index:end_index]

  segment.append(segments)

# 檢查
segment[:5]
```

```python
# 創一個 ResultP 的列表,儲存篩選後的序列

ResultP = []

for segments in segment:
    if len(segments) == 31:
        ResultP.append(segments)

# 存成FASTA檔
fasta_filename = "positive_segments.fasta"

with open(fasta_filename, "w") as fasta_file:
    for i, segments in enumerate(ResultP, start=1):
        sequence_name = f"Segment_{i}"
        fasta_file.write(f">{Ub[0][i]}_{Ub[2][i]}_{Ub[1][i]}\n")
        fasta_file.write(f"{segments}\n")

print(f"已保存到 {fasta_filename}")
```

# Negative Result

```python
npp= Ub[1]
pp_neg=[]

for i in range(len(npp)):
    pp_neg.append(npp[i] - 1)

print(pp_neg)
```

```python
collected_names = set()

for index, row in Ub.iterrows():

  names = row[0]
  sequence_neg = row[3]

#如果名字出現在collect_names,就跳出迴圈,重新來過
  if names in collected_names:
    continue
#如果不用重新來過,就收集全部的poaitive K的位置
  collected_positive_K = Ub[Ub[0] == names][1].tolist()
  collected_positive_K = [int(i)-1 for i in collected_positive_K if i]
#把處理過的名字加回set裡,才不會再找一次
  collected_names.add(names)

collected_positive_K[:5]
```

```python
modified_segment_neg = []  # 建立一個空的列表，用於儲存修改後的序列

for index, row in Ub.iterrows():
    names = row[0]
    if names in collected_names:
        # 取得對應名稱的正K位置
        collected_positive_K = Ub[Ub[0] == names][1].tolist()
        collected_positive_K = [int(i) - 1 for i in collected_positive_K if i]
        # 替換序列中的K的前後15個字母和K為1
        sequence = row[3]
        for k_index in collected_positive_K:
            start = max(0, k_index - 15)
            end = min(len(sequence), k_index + 16)
            sequence = sequence[:start] + '1' * (end - start) + sequence[end:]

        # 將處理後的序列加入到modified_segment_neg清單中
        modified_segment_neg.append(sequence)

# 现現在modified_segment_neg清單包含了所有處理後的序列

modified_segment_neg
```

```python
collected_names2 = set()
modified_segment_neg_new = []
selected_names = Ub[0].tolist()

for i in range(15000):
  names = selected_names[i]
  sequence = modified_segment_neg[i]

#如果名字出現在collect_names,就跳出迴圈,重新來過
  if names in collected_names2:
    continue
#如果不用重新來過,就收集遇到的序列
  modified_segment_neg_new.append(sequence)

#把處理過的名字加回set裡,才不會再找一次
  collected_names2.add(names)

modified_segment_neg_new[:5]

```

```python
selected_names = Ub[0].tolist()
len(selected_names)
```

```python
target_letter = 'K'  # 在字串中查找K的位置
index1 = []

negsegments = []

for sequence_neg in modified_segment_neg_new:
    # 初始化要找的位置為0
    start_find_index = 0

    # 一直找K直到該序列沒有K為止
    while True:
        # 從指定開始位置開始找K
        letter_index = sequence_neg.find(target_letter, start_find_index)

        # 找到K
        if letter_index != -1:

            index1.append(letter_index)

            # 就把K附近的序列取出來
            start_index = max(0, letter_index - 15)
            end_index = min(len(sequence_neg), letter_index + 16)
            neg_segments = sequence_neg[start_index:end_index]

            # 放到新的array裡面
            negsegments.append(neg_segments)

            # 更新找尋K的起始位置(從已經找到的K，位置+1後繼續找)
            start_find_index = letter_index + 1
        else:
            # 如果這條序列沒K了，就可以換行了
            break

# 檢查
negsegments[:20]
```

```python
index1[:20]
```

```python
# 創建一個 ResultN 的集合，用於儲存唯一的篩選後的序列
ResultN=[]
index_new=[]

for index, neg_segments in enumerate(negsegments):
    # 篩選序列長度為31，且不包含 '1'
    if len(neg_segments) == 31 and "1" not in neg_segments:
        ResultN.append(neg_segments)
        index_new.append(index1[index])

ResultN[:20]
```

```python
##### 新增的代碼區塊 #####

results = []

for i in range(len(modified_segment_neg_new)):
    sequence_neg = modified_segment_neg_new[i]
    current_name = selected_names[i]
    start_find_index = 0

    # 一直找K直到該序列沒有K為止
    while True:
        # 從指定開始位置開始找K
        letter_index = sequence_neg.find(target_letter, start_find_index)

        # 找到K
        if letter_index != -1:
            # 就把K附近的序列取出來
            start_index = max(0, letter_index - 15)
            end_index = min(len(sequence_neg), letter_index + 16)
            neg_segments = sequence_neg[start_index:end_index]

            if len(neg_segments) == 31 and "1" not in neg_segments:
                # 儲存每個`ResultN`及其對應的ID、species name和negative K position
                results.append({
                    'sequence': neg_segments,
                    'ID': current_name.split('_')[0],
                    'species_name': '_'.join(current_name.split('_')[1:]),
                    'neg_K_position': letter_index
                })

            # 更新找尋K的起始位置(從已經找到的K，位置+1後繼續找)
            start_find_index = letter_index + 1
        else:
            # 如果這條序列沒K了，就可以換行了
            break

# 输出到FASTA文件
with open("C:/Users/ddes2/Desktop/桌面/BML/output(2).fasta", 'w') as f:
    for item in results:
        header = ">{ID} {species_name}_{neg_K_position}\n".format(**item)
        f.write(header)
        f.write(item['sequence'] + '\n')

print("output(2).fasta")
```

```python
###### 新增的代碼區塊 ######

# Create a mapping from protein ID to species name
protein_to_species = {}

with open("C:/Users/ddes2/Desktop/桌面/BML/train_plant_15000.txt", "r") as file:
    for line in file:
        parts = line.split('\t')
        protein_id = parts[0]
        species_name = parts[2]
        protein_to_species[protein_id] = species_name

# Check the first few entries in the mapping
dict(list(protein_to_species.items())[:5])





# Update the fasta file with species names and save it to output(3).fasta
updated_lines = []

with open("C:/Users/ddes2/Desktop/桌面/BML/output(2).fasta", "r") as fasta_file:
    for line in fasta_file:
        if line.startswith(">"):
            protein_id = line.split()[0][1:]  # Extract protein ID without ">"
            if protein_id in protein_to_species:
                species_name = protein_to_species[protein_id]
                updated_line = f">{protein_id} {species_name}{line.split()[1]}\n"
                updated_lines.append(updated_line)
            else:
                # If protein ID is not found in the mapping, keep the original line
                updated_lines.append(line)
        else:
            updated_lines.append(line)

# Save to output(3).fasta
output_path = "C:/Users/ddes2/Desktop/桌面/BML/output(3).fasta"
with open(output_path, "w") as output_file:
    output_file.writelines(updated_lines)

output_path
```
